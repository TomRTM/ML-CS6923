```python
import pandas as pd
import numpy as np
# df = pd.read_csv('training_set_1_filter.csv',encoding = "ISO-8859-1")
df = pd.read_csv('training_data_2_csv_UTF.csv',encoding = "UTF").dropna(how='all')
```

```python
df.loc[:,'flag1'] = 1
for i in df.index:    
    if df.loc[i,'default_profile'] :
        df.loc[i,'flag1'] = 1
    else:
        df.loc[i,'flag1'] = 0
```

```python
df.loc[:,'flag2'] = 1
for i in df.index:
    if type(df.loc[i,'has_extended_profile'])==bool:
        if df.loc[i,'has_extended_profile'] :
            df.loc[i,'flag2'] = 1
        else:
            df.loc[i,'flag2'] = 0
    else:
            df.loc[i,'flag2'] = 2
```

```python
df.loc[:,'flag3'] = 1
for i in df.index:    
    if df.loc[i,'default_profile_image'] :
        df.loc[i,'flag3'] = 1
    else:
        df.loc[i,'flag3'] = 0
```

```python
df.loc[:,'flag4'] = 1
for i in df.index:    
    if df.loc[i,'verified'] :
        df.loc[i,'flag4'] = 1
    else:
        df.loc[i,'flag4'] = 0
```

```python
# Whether 'bot' is in location
df.loc[:,'flag6']=1
for i in df.index: 
    if "bot" in str(df.loc[i,'location']):
        df.loc[i,'flag6'] = 1
    else:
        df.loc[i,'flag6']=0
```

```python
#status
df.loc[:,'flag8'] = 1
for i in df.index:    
    if type(df.loc[i,'status']) == type('str') :
        if df.loc[i,'status']== 'null' or df.loc[i,'status']== 'None':
            df.loc[i,'flag8'] = 0
        elif df.loc[i,'status']=='{':
            df.loc[i,'flag8'] = 0
        else:
            df.loc[i,'flag8'] = 1
    else:
        df.loc[i,'flag8'] = 0
```

```python
# id
df['flag9'] = df[df.columns[0]].astype(np.int64)
```

```python
# description
df.loc[:,'flag10'] = 1
for i in df.index:    
    if type(df.loc[i,'description']) == type('str') :
        if df.loc[i,'description']== 'null' or df.loc[i,'description']== 'None':
#             print(df.loc[i,'description'])
            df.loc[i,'flag10'] = 1
        elif "bot" in str(df.loc[i,'description']):
#             print (df.loc[i,'description'])
            df.loc[i,'flag10'] = 1
        else:
            df.loc[i,'flag10'] = 0
    else:
        df.loc[i,'flag10'] = 0
```

```python
#bot in screen_name
df.loc[:,'flag11']=1
for i in df.index: 
    if (str(i[-3:])=='bot') or (str(i[-3:])=='Bot') or (str(i[0:3]))=='bot'or (str(i[0:3]))=='Bot':
        df.loc[i,'flag11'] = 1
    else:
        df.loc[i,'flag11'] = 0
```

```python
#status or not
df.loc[:,'flag12'] = 1
for i in df.index:    
    if "entities" in str(df.loc[i,'status']) :
        df.loc[i,'flag12'] = 0
```

```python
# create_at 
df.loc[:,'flag13']= 0
for i in df.index: 
    if "2011" in str(df.loc[i,'created_at']):
        df.loc[i,'flag13'] = 1
#         print(df.loc[i,'created_at'])    
    elif "2012" in str(df.loc[i,'created_at']):
        df.loc[i,'flag13'] = 2
#         print(df.loc[i,'created_at'])    
    elif "2013" in str(df.loc[i,'created_at']):
        df.loc[i,'flag13'] = 3
#         print(df.loc[i,'created_at'])    
    elif "2014" in str(df.loc[i,'created_at']):
        df.loc[i,'flag13'] = 4
#         print(df.loc[i,'created_at'])    
    elif "2015" in str(df.loc[i,'created_at']):
        df.loc[i,'flag13'] = 5
#         print(df.loc[i,'created_at'])
    elif "2016" in str(df.loc[i,'created_at']):
        df.loc[i,'flag13'] = 6
#         print(df.loc[i,'created_at'])
    elif "2017" in str(df.loc[i,'created_at']):
        df.loc[i,'flag13'] = 7
    else:
        df.loc[i,'flag13']=0
```

```python
# yearly tweets
df.loc[:,'flag14']= 0
for i in df.index: 
    df.loc[:,'flag14'] = df.loc[:,'statuses_count']*(8-df.loc[:,'flag13'])
```

```python
# flag7
df.loc[:,'flag7']= 0
for i in df.index: 
    df.loc[:,'flag14'] = df.loc[:,'statuses_count']*(8-df.loc[:,'flag13'])
```
